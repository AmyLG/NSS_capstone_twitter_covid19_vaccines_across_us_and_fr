```python
import tweepy as tw
import pandas as pd
import os
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = 3000 
```

```python
API_KEY = os.environ.get("API_KEY")
SECRET_KEY = os.environ.get("SECRET_KEY")
ACCESS_TOKEN = os.environ.get("ACCESS_TOKEN")
SECRET_TOKEN = os.environ.get("SECRET_TOKEN")
```

```python
# establish an instance of tweepy's API
auth = tw.OAuthHandler(API_KEY, SECRET_KEY)
auth.set_access_token(ACCESS_TOKEN, SECRET_TOKEN)
api = tw.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)

try:
    api.verify_credentials()

    print("Authentication OK")

except:

  print("Error during authentication")
```

## 1-Tweets in Southern US States:

### Getting Pfizer only tweets:

```python
# define tweets on covid vaccine in french
search_words = 'pfizer -moderna -astrazeneca -"johnson & johnson"'
new_search = search_words + "-filter:retweets"
date_since = "2021-05-15"
date_until = "2021-05-22"
country = "33.543682,-86.779633,30km" #Birmingham location lat, long
```

```python
# Download relevant tweets from search and save it in a dataset
pfizer_us_s_dataset = pd.DataFrame(columns=['Tweet Id', 'Tweet Date', 'location', 'Follower Count', 'Account Verified', 'Favorite Count', 'Retweets', 'Tweet Text'])

counter = 0
for tweet in tw.Cursor(api.search, count=5, tweet_mode='extended', q=new_search, lang="en", geocode=country, since=date_since, until=date_until).items(): #
        print('Tweet Downloaded: ', counter)
        appending_dataframe = pd.DataFrame([[tweet.id, tweet.created_at, tweet.user.location, tweet.user.followers_count, tweet.user.verified, tweet.favorite_count, tweet.retweet_count, tweet.full_text.encode('utf-8')]], columns=['Tweet Id', 'Tweet Date', 'location', 'Follower Count', 'Account Verified', 'Favorite Count', 'Retweets', 'Tweet Text'])
        pfizer_us_s_dataset = pfizer_us_s_dataset.append(appending_dataframe)
        counter+=1
        if counter >= 5:
            break
```
