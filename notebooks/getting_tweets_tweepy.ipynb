```python
import tweepy as tw
import pandas as pd
import os
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = 300
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
# Twitter API keys and tokens
API_KEY = os.environ.get("API_KEY")
SECRET_KEY = os.environ.get("SECRET_KEY")
ACCESS_TOKEN = os.environ.get("ACCESS_TOKEN")
SECRET_TOKEN = os.environ.get("SECRET_TOKEN")
```

```python
# establish an instance of tweepy's API
auth = tw.OAuthHandler(API_KEY, SECRET_KEY)
auth.set_access_token(ACCESS_TOKEN, SECRET_TOKEN)
api = tw.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)

try:
    api.verify_credentials()

    print("Authentication OK")

except:

  print("Error during authentication")
```

## 1-Tweets from Southern US States:

### Getting Pfizer only tweets:

**Birmingham, AL:**

```python
# define tweets on covid vaccine in AL, MS, GA, TN
search_words = 'pfizer -moderna -astrazeneca -"J&J jansen" -"johnson&johnson" -"johnson and johnson"'
new_search = search_words + "-filter:retweets"
date_since = "2021-05-23"
date_until = "2021-06-01"
country = "33.543682,-86.779633,350km" #Birmingham, AL location lat, long
```

```python
# Download relevant tweets from search and save it in a dataset
pfizer_us_b_dataset = pd.DataFrame(columns=['Tweet Id', 'username', 'acctdesc', 'Tweet Date', 'location', 'Friends Count','Follower Count', 'Account Verified', 'Favorite Count', 'Retweets', 'Tweet Text', 'hashtags'])

counter = 0
for tweet in tw.Cursor(api.search, count=10000, tweet_mode='extended', q=new_search, lang="en", geocode=country, since=date_since, until=date_until).items(): 
        print('Tweet Downloaded: ', counter)
        appending_dataframe = pd.DataFrame([[tweet.id, tweet.user.screen_name, tweet.user.description, tweet.created_at, tweet.user.location, tweet.user.friends_count, tweet.user.followers_count, tweet.user.verified, tweet.favorite_count, tweet.retweet_count, tweet.full_text.encode('utf-8'), tweet.entities['hashtags']]], columns=['Tweet Id', 'username', 'acctdesc', 'Tweet Date', 'location','Friends Count', 'Follower Count', 'Account Verified', 'Favorite Count', 'Retweets', 'Tweet Text', 'hashtags'])
        pfizer_us_b_dataset = pfizer_us_b_dataset.append(appending_dataframe)
        counter+=1
        if counter >= 10000:
            break
```

```python
# Download relevant tweets from search and save it in a dataset
tweets_fr_j_dataset = pd.DataFrame(columns=['Tweet Id', 'username', 'acctdesc', 'Tweet Date', 'location', 'Friends Count','Follower Count', 'Account Verified', 'Favorite Count', 'Retweets', 'Tweet Text', 'hashtags'])

counter = 0
for tweet in tw.Cursor(api.search, count=10000, tweet_mode='extended', q=new_search, lang="fr", geocode=country, since=date_since, until=date_until).items(): 
        print('Tweet Downloaded: ', counter)
        appending_dataframe = pd.DataFrame([[tweet.id, tweet.user.screen_name, tweet.user.description, tweet.created_at, tweet.user.location, tweet.user.friends_count, tweet.user.followers_count, tweet.user.verified, tweet.favorite_count, tweet.retweet_count, tweet.full_text.encode('utf-8'), tweet.entities['hashtags']]], columns=['Tweet Id', 'username', 'acctdesc', 'Tweet Date', 'location','Friends Count', 'Follower Count', 'Account Verified', 'Favorite Count', 'Retweets', 'Tweet Text', 'hashtags'])
        tweets_fr_j_dataset = tweets_fr_j_dataset.append(appending_dataframe)
        counter+=1
        if counter >= 10000:
            break
```

```python
tweets_fr_jandj_dataset = tweets_fr_j_dataset
```

```python
# Save as csv file
tweets_fr_jandj_dataset.to_csv('../data/tweets_fr/tweets_fr_jandj_2021_05_29_to_05_22.csv', index=False)
```
